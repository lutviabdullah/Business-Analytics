```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
data = {
    'Country': ['Thailand', 'Indonesia', 'Singapore'] * 3,
    'Category': ['Natural', 'Halal', 'Anti-Aging'] * 3,
    'Demand_Index': [78, 82, 68, 65, 70, 75, 45, 62, 55 ]
}
df = pd.DataFrame(data)
```

```python
# Visualization
plt.figure(figsize=(10,6))
for country in df['Country'].unique():
    subset = df[df['Country'] == country]
    plt.plot(subset['Category'], subset['Demand_Index'], marker='o', label=country)

plt.title("Skincare Demand Drivers in SEA (2024)")
plt.ylabel("Consumer Priority Index (%)")
plt.ylim(40, 90)
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
# data
countries = ['Indonesia', 'Thailand', 'Singapore']
halal_share = [55, 3, 5]  # % of total skincare market
growth_rates = [22, 12, 8]  # YoY growth %
```

```python
fig, ax1 = plt.subplots(figsize=(10,6))

# Bar plot for market share
bars = ax1.bar(countries, halal_share, color=['#FF6B6B', '#4ECDC4', '#556270'], alpha=0.8)
ax1.set_ylabel('% of Total Skincare Market', fontsize=12)
ax1.set_ylim(0, 60)

# Line plot for growth rates
ax2 = ax1.twinx()
ax2.plot(countries, growth_rates, 'o-', color='#C44D58', linewidth=3, markersize=10)
ax2.set_ylabel('YoY Growth Rate (%)', fontsize=12)
ax2.set_ylim(0, 25)


# Annotations
for i, v in enumerate(halal_share):
    ax1.text(i, v+1.5, f"{v}%", ha='center', fontweight='bold')
for i, v in enumerate(growth_rates):
    ax2.text(i, v+1, f"+{v}%", ha='center', fontweight='bold')

plt.title("Halal Skincare Market Comparison (2024)", fontsize=14)
plt.grid(axis='y', alpha=0.3)
fig.tight_layout()
plt.show()
```
