```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Load synthetic dataset (replace with real consumer data)
data = pd.DataFrame({
    'Age': [25, 32, 45, 28, 50, 33],
    'Income': [45000, 75000, 120000, 55000, 95000, 110000],
    'Sustainability_Interest': [8, 9, 7, 9, 6, 10],  # 1-10 scale
    'Price_Sensitivity': [6, 4, 3, 5, 7, 2],
    'Packaging_Recycling_Rate': [0.3, 0.8, 0.5, 0.9, 0.4, 0.95],
    'Brand_Loyalty': [5, 8, 6, 9, 4, 10]
})

# Sustainability segmentation
kmeans = KMeans(n_clusters=3)
data['Segment'] = kmeans.fit_predict(data[['Sustainability_Interest', 'Packaging_Recycling_Rate']])

# Visualization
plt.figure(figsize=(12, 5))
sns.scatterplot(
    x='Sustainability_Interest',
    y='Packaging_Recycling_Rate',
    hue='Segment',
    size='Income',
    data=data,
    palette='viridis',
    sizes=(50, 200)
)
plt.title("Consumer Sustainability Segmentation", fontsize=14)
plt.xlabel("Sustainability Interest (1-10)", fontsize=12)
plt.ylabel("Packaging Recycling Rate", fontsize=12)
plt.grid(alpha=0.2)
plt.show()

# Key metrics correlation
corr_matrix = data.corr()
sns.heatmap(corr_matrix[['Sustainability_Interest']],
            annot=True,
            cmap='coolwarm',
            vmin=-1,
            vmax=1)
plt.title("Sustainability Interest Correlations", fontsize=14)
plt.show()
```
